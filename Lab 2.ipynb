```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv") 
# https://www.kaggle.com/datasets/xcufx0qc2os1/the-movie-database-top-10000-movies-all-time?resource=download

ratings = df['vote_average']

plt.hist(ratings)
plt.title('Movie Rating Distribution')
plt.xlabel('Avg. Rating Out of 10')
plt.ylabel('Number of Movies with Rating')
plt.show()
```
